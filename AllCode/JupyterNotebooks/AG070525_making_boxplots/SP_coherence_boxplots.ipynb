```python
import numpy as np
import matplotlib.pyplot as plt

def plot_coherence_bands(in_dict, out_dict, bands=None, width=0.35):
    # If no explicit order given, use the dict's insertion order
    if bands is None:
        bands = list(in_dict.keys())
    n = len(bands)
    x = np.arange(n)

    # Collect data in the specified order
    data_in  = [in_dict[b]  for b in bands]
    data_out = [out_dict[b] for b in bands]

    fig, ax = plt.subplots(figsize=(10, 7))

    # Boxplots: In at x-width/2, Out at x+width/2
    bp_in  = ax.boxplot(data_in,
                        positions=x - width/2,
                        widths=width,
                        notch=False,
                        patch_artist=True,
                        boxprops=dict(facecolor='C0', alpha=0.6),
                        medianprops=dict(color='k'))
    bp_out = ax.boxplot(data_out,
                        positions=x + width/2,
                        widths=width,
                        notch=False,
                        patch_artist=True,
                        boxprops=dict(facecolor='C1', alpha=0.6),
                        medianprops=dict(color='k'))

    ax.set_xticks(x)
    ax.set_xticklabels(bands)
    ax.set_xlabel('Frequency Band')
    ax.set_ylabel('Coherence')
    ax.set_title('Coherence by Band (In vs Out)')

    # Legend
    ax.legend([bp_in["boxes"][0], bp_out["boxes"][0]],
              ['In', 'Out'],
              loc='upper right',
              fontsize=16)

    plt.tight_layout()
    plt.show()
```

```python
plot_coherence_bands(coh_in, coh_out)
```

```python
from scipy.stats import ttest_rel

bands = ['delta', 'theta', 'alpha', 'beta', 'gamma']

# Ensure the order is consistent
coh_in_arr = np.array([coh_in[band] for band in bands])   # shape (n_bands, n_datasets)
coh_out_arr = np.array([coh_out[band] for band in bands]) # shape (n_bands, n_datasets)

t_stats = []
p_values = []

for i, band in enumerate(bands):
    t_stat, p_val = ttest_rel(coh_in_arr[i], coh_out_arr[i])
    t_stats.append(t_stat)
    p_values.append(p_val)
    print(f"{band}: t = {t_stat:.3f}, p-value = {p_val:.3g}")
```
