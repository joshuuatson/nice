```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Create a sample continuous signal (sine + noise)
np.random.seed(0)
t = np.linspace(0, 1, 200)
ts = 0.001 * np.sin(2 * np.pi * 5 * t) + 0.002 * np.random.randn(len(t))

# 2) Discretization into B=5 equipopulated bins
def discretize(ts, n_states=10):
    qs = np.linspace(0, 1, n_states + 1)
    edges = np.quantile(ts, qs)
    return np.digitize(ts, edges[1:-1], right=True).astype(int)

states = discretize(ts, n_states=10)

# 3) Plot original vs discretized
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
ax1.plot(t, ts)
ax1.set_title("Original Continuous Signal")
ax1.set_ylabel("Amplitude")

ax2.step(t, states, where='mid')
ax2.set_title("Discretized Signal (5 States)")
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("State")

plt.tight_layout()
plt.show()
```
